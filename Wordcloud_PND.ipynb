```python
import PyPDF2

with open('/PND.pdf', 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text_bytes = b''

    for page_num in range(20, 320):
        page = pdf_reader.pages[page_num]
        text_bytes += page.extract_text().encode('utf-8')

    text = text_bytes.decode('utf-8')




```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import nltk

# Descargar las stopwords de NLTK para español si no están disponibles
nltk.download('stopwords')

# Definir la lista de palabras a excluir
exclusion_list = list(STOPWORDS) + ["acciones","Plan","lo","permitan","Nacional","desarrollo","manera","entre","fortalecerá","sobre",
                                    "todo","plan Nacional","l","uso","mayor", "otro", "diferentes",
                                    "partir", "mismo","marco","mismo","marco","fin","1","uso","implementará","además","desde","este","una"
                                    ,"país","es","sus","esto","través","esta","así","como","más","o","por","su","a", "b", "c", "d", "e", "f", 
                                    "g", "h", "i",'gobierno', 'estado', "para", "del", "con", 'de', 'la', 'que', 'el', 'en', 'y', 'a', 'los', "garantía",
                                    "las", "ante", "lo", "al", "un", "uno", "unos", "con" 'del', "otros", "estos", "s", "promoverá", "hacia", "fuente",
                                    'se', '[,.:;\\<\\=\\>\\?\\@\\[\\\\\\]\\^_\\`\\{\\|\\}\\~]', "_"]

# Crear el objeto WordCloud y especificar las palabras a excluir
wordcloud = WordCloud(stopwords=exclusion_list, width=1000, height=800, max_words=75, background_color="white", contour_width=1, contour_color='steelblue')

# Generar la nube de palabras
wordcloud.generate(text)

# Configurar la figura de matplotlib y mostrar la nube de palabras
plt.figure(figsize=(50, 40))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()


```
